```python
def solve():
    path = "C:/Users/zosia/OneDrive/Pulpit/day7.txt"
    with open(path, 'r') as f:
        grid = [line.rstrip('\n') for line in f]

    active_beams = set()
    for col, char in enumerate(grid[0]):
        if char == 'S':
            active_beams.add(col)
            break

    total_splits = 0
    height = len(grid)
    width = max(len(row) for row in grid)

    for r in range(height):
        next_beams = set()
        for c in active_beams:
            if 0 <= c < width:
                char = grid[r][c] if c < len(grid[r]) else '.'
                
                if char == '^':
                    total_splits += 1
                    next_beams.add(c - 1)
                    next_beams.add(c + 1)
                else:
                    next_beams.add(c)
        active_beams = next_beams

    print(total_splits)

if __name__ == "__main__":
    solve()
```

```python
from collections import defaultdict

def solve():
    path = "C:/Users/zosia/OneDrive/Pulpit/day7.txt"
    with open(path, 'r') as f:
        grid = [line.rstrip('\n') for line in f]

    current_timelines = defaultdict(int)
    for col, char in enumerate(grid[0]):
        if char == 'S':
            current_timelines[col] = 1
            break

    height = len(grid)
    width = max(len(row) for row in grid)

    for r in range(height):
        next_timelines = defaultdict(int)
        for c, count in current_timelines.items():
            if 0 <= c < width:
                char = grid[r][c] if c < len(grid[r]) else '.'
                
                if char == '^':
                    if 0 <= c - 1 < width:
                        next_timelines[c - 1] += count
                    if 0 <= c + 1 < width:
                        next_timelines[c + 1] += count
                else:
                    next_timelines[c] += count
        
        current_timelines = next_timelines

    print(sum(current_timelines.values()))

if __name__ == "__main__":
    solve()
```
