# Day 3 - Advent of Code

```python
path = "C:/Users/zosia/OneDrive/Pulpit/day3.txt"

total = 0
with open(path, 'r') as f:
    for line in f:
        bank = line.strip()
        if not bank:
            continue
            
        max_joltage = 0
        
        for i in range(len(bank)):
            for j in range(i + 1, len(bank)):
                digit1 = bank[i]
                digit2 = bank[j]
                joltage = int(digit1 + digit2)
                if joltage > max_joltage:
                    max_joltage = joltage
        
        total += max_joltage

print(total)
```

```python
def find_max_12_digit_number(bank):
    n = len(bank)
    need = 12
    result = []
    start = 0
    
    for pos in range(12):
        positions_left = n - start
        can_skip = positions_left - need
        
        max_digit = -1
        max_pos = -1
        
        for i in range(start, start + can_skip + 1):
            if int(bank[i]) > max_digit:
                max_digit = int(bank[i])
                max_pos = i
        
        result.append(str(max_digit))
        start = max_pos + 1
        need -= 1
    
    return int(''.join(result))

total = 0
with open(path, 'r') as f:
    for line in f:
        bank = line.strip()
        if not bank or len(bank) < 12:
            continue
        
        max_joltage = find_max_12_digit_number(bank)
        total += max_joltage

print(total)
```
