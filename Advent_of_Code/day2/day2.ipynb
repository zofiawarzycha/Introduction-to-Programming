# Day 2 - Advent of Code

```python
path = "C:/Users/zosia/OneDrive/Pulpit/day2.txt"

def is_invalid_id(n):
    s = str(n)
    l = len(s)
    if l % 2 != 0:
        return False
    half = l // 2
    return s[:half] == s[half:]

with open(path) as f:
    line = f.read().strip()
ranges = line.split(',')

invalid_ids = []

for r in ranges:
    if '-' not in r:
        continue
    start, end = map(int, r.split('-'))
    for num in range(start, end + 1):
        if is_invalid_id(num):
            invalid_ids.append(num)

print(sum(invalid_ids))
```

```python
def is_invalid_id_part2(n):
    s = str(n)
    length = len(s)
    for sub_len in range(1, length // 2 + 1):
        if length % sub_len == 0:
            sub = s[:sub_len]
            if sub * (length // sub_len) == s:
                return True
    return False

with open(path) as f:
    line = f.read().strip()
ranges = line.split(',')

invalid_ids = []

for r in ranges:
    if '-' not in r:
        continue
    start, end = map(int, r.split('-'))
    for num in range(start, end + 1):
        if is_invalid_id_part2(num):
            invalid_ids.append(num)

print(sum(invalid_ids))
```
