```python
def solve_cafeteria_puzzle():
    file_path = "C:/Users/zosia/OneDrive/Pulpit/day5.txt"
    ranges = []
    ids = []
    parsing_ranges = True

    try:
        with open(file_path, 'r') as f:
            for line in f:
                line = line.strip()
                
                if line.startswith('[source'):
                    line = line.split(']')[-1].strip()

                if not line:
                    if parsing_ranges and len(ranges) > 0:
                        parsing_ranges = False
                    continue

                if parsing_ranges:
                    if '-' in line:
                        parts = line.split('-')
                        ranges.append((int(parts[0]), int(parts[1])))
                    else:
                        parsing_ranges = False
                        ids.append(int(line))
                else:
                    ids.append(int(line))

        fresh_count = 0
        
        for ingredient_id in ids:
            for start, end in ranges:
                if start <= ingredient_id <= end:
                    fresh_count += 1
                    break

        print(fresh_count)

    except FileNotFoundError:
        print("File not found.")

if __name__ == "__main__":
    solve_cafeteria_puzzle()
```

```python
def solve_cafeteria_part_two():
    file_path = "C:/Users/zosia/OneDrive/Pulpit/day5.txt"
    ranges = []

    try:
        with open(file_path, 'r') as f:
            for line in f:
                line = line.strip()
                if line.startswith('[source'):
                    line = line.split(']')[-1].strip()
                
                if not line:
                    break
                
                if '-' in line:
                    parts = line.split('-')
                    start, end = int(parts[0]), int(parts[1])
                    ranges.append((start, end))

        if not ranges:
            print(0)
            return

        ranges.sort(key=lambda x: x[0])

        merged_ranges = []
        curr_start, curr_end = ranges[0]

        for i in range(1, len(ranges)):
            next_start, next_end = ranges[i]

            if next_start <= curr_end + 1:
                curr_end = max(curr_end, next_end)
            else:
                merged_ranges.append((curr_start, curr_end))
                curr_start, curr_end = next_start, next_end
        
        merged_ranges.append((curr_start, curr_end))

        total_fresh = 0
        for start, end in merged_ranges:
            total_fresh += (end - start + 1)

        print(total_fresh)

    except FileNotFoundError:
        print("File not found.")

if __name__ == "__main__":
    solve_cafeteria_part_two()
```
