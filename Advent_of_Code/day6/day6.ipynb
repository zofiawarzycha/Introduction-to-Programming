```python
import sys

def solve():
    file_path = "C:/Users/zosia/OneDrive/Pulpit/day6.txt"
    
    try:
        with open(file_path, 'r') as f:
            lines = f.read().splitlines()
    except FileNotFoundError:
        print(f"Error: Could not find file at {file_path}")
        return

    if not lines:
        print(0)
        return

    max_width = max(len(line) for line in lines)
    grid = [line.ljust(max_width) for line in lines]
    height = len(grid)
    
    grand_total = 0
    col = 0
    
    while col < max_width:
        is_empty = True
        for row in range(height):
            if grid[row][col] != ' ':
                is_empty = False
                break
        
        if is_empty:
            col += 1
            continue
            
        start_col = col
        while col < max_width:
            col_is_separator = True
            for row in range(height):
                if grid[row][col] != ' ':
                    col_is_separator = False
                    break
            
            if col_is_separator:
                break
            col += 1
        
        end_col = col
        
        problem_numbers = []
        operator = None

        for c in range(end_col - 1, start_col - 1, -1):
            num_str = ""
            for r in range(height):
                char = grid[r][c]
                if char.isdigit():
                    num_str += char
                elif char in ('+', '*'):
                    operator = char
            
            if num_str:
                problem_numbers.append(int(num_str))

        if operator == '+':
            result = sum(problem_numbers)
        elif operator == '*':
            result = 1
            for n in problem_numbers:
                result *= n
        else:
            result = 0
            
        grand_total += result

    print(grand_total)

if __name__ == "__main__":
    solve()
```
