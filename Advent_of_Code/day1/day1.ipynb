```python
path = "C:/Users/zosia/OneDrive/Pulpit/day1.txt"

def solve_safe_combinations(input_file):
    position = 50
    zero_count = 0
    
    try:
        with open(input_file, 'r', encoding='utf-8') as file:
            lines = file.readlines()
        
        for line in lines:
            line = line.strip()
            if not line:
                continue
                
            direction = line[0]
            distance = int(line[1:])
            
            if direction == 'R':
                for _ in range(distance):
                    position = (position + 1) % 100
                    if position == 0:
                        zero_count += 1
            elif direction == 'L':
                for _ in range(distance):
                    position = (position - 1) % 100
                    if position == 0:
                        zero_count += 1
        
        return zero_count
        
    except FileNotFoundError:
        print(f"Error: Could not find file '{input_file}'")
        return None
    except ValueError:
        print("Error parsing rotation")
        return None
    except Exception:
        print("Unexpected error")
        return None

if __name__ == "__main__":
    path = "C:/Users/zosia/OneDrive/Pulpit/day1.txt"
    result = solve_safe_combinations(path)
    
    if result is not None:
        print(f"Password: {result}")
    else:
        print("Double-check your file path!")
```
